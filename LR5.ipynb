### Хід роботи

```python
# ==========================================
# Завдання 5 (Варіант 1)
# ==========================================
import numpy as np
import matplotlib.pyplot as plt

# Вхідні дані
t = np.array([0, 0.02, 0.04, 0.06, 0.08, 0.1, 0.12, 0.14, 0.16, 0.18, 0.2])
f = np.array([3, 3.137, 3.267, 3.391, 3.509, 3.621, 3.727, 3.826, 3.919, 4.007, 4.088])

h = 0.02
n = len(t) - 1

# ==========================================
# 1. Розрахунок першої похідної
# ==========================================
df1 = np.zeros(len(t))

# Внутрішні точки
for i in range(1, n):
    df1[i] = (f[i + 1] - f[i - 1]) / (2 * h)

# Крайні точки
df1[0] = (-3*f[0] + 4*f[1] - f[2]) / (2 * h)
df1[n] = (f[n-2] - 4*f[n-1] + 3*f[n]) / (2 * h)

# ==========================================
# 2. Розрахунок другої похідної
# ==========================================
df2 = np.zeros(len(t))

# Внутрішні точки
for i in range(1, n):
    df2[i] = (f[i - 1] - 2*f[i] + f[i + 1]) / (h**2)

# Крайні точки
df2[0] = (2*f[0] - 5*f[1] + 4*f[2] - f[3]) / (h**2)
df2[n] = (-f[n-3] + 4*f[n-2] - 5*f[n-1] + 2*f[n]) / (h**2)

# ==========================================
# 3. Побудова графіків (два окремих)
# ==========================================
fig, axs = plt.subplots(2, 1, figsize=(10, 8)) # 2 рядки, 1 колонка

# Верхній графік: Перша похідна
axs[0].plot(t, df1, 'ro-', label="Перша похідна f'(t)")
axs[0].set_title("Завдання 5 (Варіант 1): Чисельне диференціювання")
axs[0].set_ylabel("f'(t)")
axs[0].grid(True)
axs[0].legend()

# Нижній графік: Друга похідна
axs[1].plot(t, df2, 'bs-', label="Друга похідна f''(t)")
axs[1].set_xlabel("Час (t)")
axs[1].set_ylabel("f''(t)")
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```
